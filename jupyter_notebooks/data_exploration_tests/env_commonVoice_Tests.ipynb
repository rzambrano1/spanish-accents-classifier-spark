**Test Librosa**

librosa is a python package for music and audio analysis. It provides the building blocks necessary to create music information retrieval systems

```python
# Beat tracking example
import librosa
```

```python
# 1. Get the file path to an included audio example
filename = librosa.example('nutcracker')
```

```python
# 2. Load the audio as a waveform `y`
#    Store the sampling rate as `sr`
y, sr = librosa.load(filename)
```

```python
# 3. Run the default beat tracker
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)
```

```python
print('Estimated tempo: {:.2f} beats per minute'.format(tempo))
```

```python
# 4. Convert the frame indices of beat events into timestamps
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
```

**Testing PyAudio**

```python
"""PyAudio Example: Play a wave file."""

import wave
import sys
import pyaudio
```

```python
CHUNK = 1024

if len(sys.argv) < 2:
    print(f'Plays a wave file. Usage: {sys.argv[0]} filename.wav')
    sys.exit(-1)
```

```python
sys.argv
```

```python
import os
os.getcwd()
```

```python
#wf = wave.open("PinkPanther30.wav", 'rb')
wf = wave.open("english.wav", 'rb')
```

```python
# Instantiate PyAudio and initialize PortAudio system resources (1)
p = pyaudio.PyAudio()
```

```python
# Get the number of audio I/O devices
devices = p.get_device_count()

# Print the total number of devices
print(f'Total number of devices: {devices}')
```

```python
p.get_default_input_device_info()
```

```python
p.get_default_output_device_info()
```

```python
# Open stream (2)
stream = p.open(format=p.get_format_from_width(wf.getsampwidth()),
                channels=wf.getnchannels(),
                rate=wf.getframerate(),
                output=True)
```

```python
# Play samples from the wave file (3)
while len(data := wf.readframes(CHUNK)):  # Requires Python 3.8+ for :=
    stream.write(data)
```

```python
# Close stream (4)
stream.close()
```

```python
# Release PortAudio system resources (5)
p.terminate()
```

The IPython Audio widget accepts raw numpy data as audio signals. This means we can synthesize signals directly and play them back in the browser.

**Testing SpeechRecognition**

```python
import speech_recognition as sr
```

```python
# obtain path to "english.wav" in the same folder as this script
import os
from os import path
# AUDIO_FILE = path.join(path.dirname(path.realpath("C:/Users/rzamb/Documents/Desktop/UMD/651_Big_Data2/finalProjectTest2/english.wav")),"english.wav")
# AUDIO_FILE = path.join(path.dirname(path.realpath(__file__)), "french.aiff")
# AUDIO_FILE = path.join(path.dirname(path.realpath(__file__)), "chinese.flac")
AUDIO_FILE = "english.wav"
```

```python
AUDIO_FILE
```

```python
# use the audio file as the audio source
r = sr.Recognizer()
with sr.AudioFile(AUDIO_FILE) as source:
    audio = r.record(source)  # read the entire audio file
```

```python
# recognize speech using Sphinx
try:
    print("Sphinx thinks you said " + r.recognize_sphinx(audio))
except sr.UnknownValueError:
    print("Sphinx could not understand audio")
except sr.RequestError as e:
    print("Sphinx error; {0}".format(e))
```

```python
# recognize speech using Google Speech Recognition
try:
    # for testing purposes, we're just using the default API key
    # to use another API key, use `r.recognize_google(audio, key="GOOGLE_SPEECH_RECOGNITION_API_KEY")`
    # instead of `r.recognize_google(audio)`
    print("Google Speech Recognition thinks you said " + r.recognize_google(audio))
except sr.UnknownValueError:
    print("Google Speech Recognition could not understand audio")
except sr.RequestError as e:
    print("Could not request results from Google Speech Recognition service; {0}".format(e))
```
